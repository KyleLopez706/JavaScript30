```
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

data = load_iris()

X = data.data
y = data.target

from sklearn.model_selection import train_test_split
X_Train, X_Test, y_Train, y_Test = train_test_split(X, y, test_size = 0.2, random_state = 0)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)

knn. fit(X_Train, y_Train)

predictions = knn.predict(X_Test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_Test, predictions)
print("Accuracy :", accuracy)

from sklearn.metrics import classification_report
print("classification report")
print(classification_report(y_Test, predictions))

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_Test, predictions)

print("Confusion matrix :")
print(cm)

plt.figure(figsize=(8, 6))
for i in range(3):
  plt.scatter(X_Test[:, 0][y_Test == i], X_Test[:, 1][y_Test == i], cmap = 'viridis')
  plt.scatter(X_Test[:, 0], X_Test[:, 1], c = predictions, cmap = 'viridis', alpha = 0.8)
  plt.xlabel(data.feature_names[0])
  plt.ylabel(data.feature_names[1])
  plt.title("Scatter Plot with Predicted Classes")
  plt.legend()
  plt.show()

  plt.figure(figsize = (8,6))
plt.imshow(cm, cmap = 'Blues')
plt.colorbar()
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.xticks(ticks = range(3), labels = data.target_names)
plt.xticks(ticks = range(3), labels = data.target_names)
plt.show()
```

# Load Data, Seperate features and target, Splitting data for training and testing





```
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

data = load_iris()

X = data.data
y = data.target

from sklearn.model_selection import train_test_split
X_Train, X_Test, y_Train, y_Test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Creata KNeighborsClassifier object, Train model using training data, Use the model to predict the test set

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)

knn. fit(X_Train, y_Train)

predictions = knn.predict(X_Test)
```

# Check accuracy, Evaluate model performance

```
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_Test, predictions)
print("Accuracy :", accuracy)

from sklearn.metrics import classification_report
print("classification report")
print(classification_report(y_Test, predictions))
```

# Making confusion matrix

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_Test, predictions)

print("Confusion matrix :")
print(cm)
```

# Visualization of Scatter plot with predicted classes

```
plt.figure(figsize=(8, 6))
for i in range(3):
  plt.scatter(X_Test[:, 0][y_Test == i], X_Test[:, 1][y_Test == i], cmap = 'viridis')
  plt.scatter(X_Test[:, 0], X_Test[:, 1], c = predictions, cmap = 'viridis', alpha = 0.8)
  plt.xlabel(data.feature_names[0])
  plt.ylabel(data.feature_names[1])
  plt.title("Scatter Plot with Predicted Classes")
  plt.legend()
  plt.show()
```

# Confusion matrix visualization

```
plt.figure(figsize = (8,6))
plt.imshow(cm, cmap = 'Blues')
plt.colorbar()
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.xticks(ticks = range(3), labels = data.target_names)
plt.xticks(ticks = range(3), labels = data.target_names)
plt.show()
```
